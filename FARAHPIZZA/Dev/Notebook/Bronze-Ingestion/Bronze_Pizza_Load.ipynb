```python
# Convert date and time columns to proper types
df['order_date'] = pd.to_datetime(df['order_date'], format="%d-%m-%Y").dt.date
df['order_time'] = pd.to_datetime(df['order_time'], format="%H:%M:%S").dt.time


df.info()
```

```python

# Step 3 - Generate CREATE TABLE based on DataFrame columns/dtypes
def generate_create_table_sql(schema, table_name, df):
    type_mapping = {
        "int64": "INT",
        "float64": "FLOAT",
        "object": "VARCHAR(255)",   # Most CSV columns are strings
        "datetime64[ns]": "DATETIME"
    }

    columns = []
    for col, dtype in df.dtypes.items():
        sql_type = type_mapping.get(str(dtype), "VARCHAR(255)")
        columns.append(f"[{col}] {sql_type}")

    columns_sql = ",\n".join(columns)

    create_table_sql = f"""
    IF OBJECT_ID('{schema}.{table_name}', 'U') IS NULL
    BEGIN
        CREATE TABLE {schema}.{table_name} (
            {columns_sql}
        )
    END
    """
    return create_table_sql

# Generate and execute the SQL
create_table_sql = generate_create_table_sql("bronze", "pizza_sales", df)
cursor.execute(create_table_sql)
conn.commit()
print("Table bronze.pizza_sales created (or already exists).")

# Step 4 - Truncate existing data before inserting new data - optional
cursor.execute("TRUNCATE TABLE bronze.pizza_sales")
conn.commit()
print("Table bronze.pizza_sales truncated.")

# Step 5 - Optionally, insert data into the table
# If you want to bulk insert the data, do this:
for index, row in df.iterrows(): 
    values = tuple(row)
    placeholders = ','.join('?' * len(row))
    sql = f"INSERT INTO bronze.pizza_sales VALUES ({placeholders})"
    cursor.execute(sql, values)

conn.commit()
print("Data inserted into bronze.pizza_sales.")
```
